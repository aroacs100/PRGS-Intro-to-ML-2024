```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate random data
x = np.linspace(0, 10, 100)
y1 = np.sin(x) + np.random.normal(0, 0.1, 100)  # Sine wave with noise
y2 = np.cos(x) + np.random.normal(0, 0.1, 100)  # Cosine wave with noise
data = np.random.randn(1000)  # Normally distributed data

# Create a figure and subplots
plt.figure(figsize=(12, 8))

# Scatter plot
plt.subplot(2, 2, 1)
plt.scatter(x, y1, color='blue', label='Sin(x) + noise')
plt.scatter(x, y2, color='red', label='Cos(x) + noise')
plt.title('Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Line plot
plt.subplot(2, 2, 2)
plt.plot(x, y1, color='blue', label='Sin(x) + noise')
plt.plot(x, y2, color='red', label='Cos(x) + noise')
plt.title('Line Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Histogram
plt.subplot(2, 2, 3)
plt.hist(data, bins=30, color='green', edgecolor='black', alpha=0.7)
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Combined line and scatter plot
plt.subplot(2, 2, 4)
plt.plot(x, y1, color='blue', label='Sin(x) + noise')
plt.scatter(x, y1, color='blue')
plt.plot(x, y2, color='red', label='Cos(x) + noise')
plt.scatter(x, y2, color='red')
plt.title('Combined Line and Scatter Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
